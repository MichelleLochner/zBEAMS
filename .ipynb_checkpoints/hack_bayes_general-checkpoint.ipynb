```python
import numpy as np
import scipy as sp
from scipy.integrate import quad
import matplotlib.pyplot as plt
from scipy.interpolate import griddata, splrep, splev, bisplrep, bisplev

%matplotlib inline
```

# Part 1
## Fitting the data with MCMC

```python
#load Union2.1 data
dados=np.loadtxt('SCPUnion2.1_mu_vs_z.txt',usecols=[1,2,3])
z=dados[:,0]
mu_o=dados[:,1]
err_mu=dados[:,2]
```

```python
#data spead with error bars
plt.figure(figsize = (12,6))
plt.errorbar(z,mu_o,yerr=[err_mu,err_mu],fmt='bo')
plt.ylabel(r'$\mu$')
plt.xlabel(r'$z$')
plt.show()
```

```python
# general functions

def H(z,OM,H0,w):
    if w==-1:
        return H0*sp.sqrt(OM*(1+z)**3+(1-OM))
    else:
        return H0*sp.sqrt(OM*(1+z)**3+(1-OM)*(1+z)**(3*(w+1)))

c=2.99792e5

def dL(z,OM,H0,w):
    return (1+z)*quad(lambda x:c/H(x,OM,H0,w),0,z)[0]

def mu(z,OM,H0,w):
    return 5*sp.log10(dL(z,OM,H0,w))+25
    
def chi2(mu_o,mu_t,sigma):
    return sum((mu_o-mu_t)**2/sigma**2)
  
#make computing mu faster

def interpoler_mu(OM,H0,w,z_data):
    z_spl = np.linspace(min(z_data), max(z_data),100)
    mu_spl = np.array([mu(z_spli,OM,H0,w) for z_spli in z_spl])
    mu_spl_rep = splrep(z_spl, mu_spl)
    mu_theory=splev(z_data, mu_spl_rep)
    return mu_theory
```

```python
#check %error in using interpoler
plt.plot(z,abs(np.array([mu(zi,0.3,70,-1) for zi in z])-interpoler_mu(0.3,70,-1,z))/np.array([mu(zi,0.3,70,-1) for zi in z])*100,'ko')
plt.show()
```

```python
#MCMC stuff
def prior_om(x):
    if x>=0 and x<=1:
        return 1
    else:
        return 0
    
def prior_w(x):
    if x>=-4 and x<=1:
        return 1
    else:
        return 0
    
def prior_H0(x):
    if x>=50 and x<=90:
        return 1
    else:
        return 0
    
def MCMC_SNe(N,z,mu_o,err_mu):
    #create arrays
    o_m=np.zeros(N)
    w=np.zeros(N)
    H0=np.zeros(N)
    chi2_mcmc=np.zeros(N)
    
    #set starting point
    o_m[0]=np.random.rand()
    w[0]=np.random.rand()-1.5
    H0[0]=np.random.rand()*20+60
    
    #first point
    mu_t=interpoler_mu(o_m[0],H0[0],w[0],z)
    chi2_mcmc[0]=chi2(mu_o,mu_t,err_mu)
    
    #set default variables of the current point
    o_m_p=o_m[0]
    w_p=w[0]
    H0_p=H0[0]
    chi2_p=chi2_mcmc[0]
    
    #dummy var for acceptance rate
    accept=0
    for i in range(N-1):
        if i%5000==0:
            print(accept)
            
        #get new point
        o_m[i+1]=o_m_p+np.random.randn()*0.04
        w[i+1]=w_p+np.random.randn()*0.04
        H0[i+1]=H0_p+np.random.randn()*0.5
        
        #compute chi2
        mu_t=interpoler_mu(o_m[i+1],H0[i+1],w[i+1],z)
        chi2_mcmc[i+1]=chi2(mu_o,mu_t,err_mu)
        
        if prior_om(o_m[i+1])==0 or prior_w(w[i+1])==0 or prior_H0(H0[i+1])==0:
            r=0.0
            next
        else:
            #if priors not flat they need to be included after the exp
            r=sp.exp(-(chi2_mcmc[i+1]-chi2_p)/2)
        
        #accept or not
        teste=np.random.rand()
        if r>teste:
            o_m_p=o_m[i+1]
            w_p=w[i+1]
            H0_p=H0[i+1]
            chi2_p=chi2_mcmc[i+1]
            accept+=1    
    acceptr=accept/N*100
        
    return o_m,w,H0,chi2_mcmc,acceptr
```

```python
N=100000
```

```python
#get the mcmc result

o_m,w,H0,chi2_mcmc,acptr=MCMC_SNe(N,z,mu_o,err_mu)
print(acptr)
```

```python
#save data
np.savetxt('mcmc_SNe_Om_w_h0_data3.txt',np.c_[o_m,w,H0,chi2_mcmc],fmt=['%2.3f','%2.3f','%2.3f','%2.3e'],\
           header='\n'.join(["acceptance rate %.2f" %acptr, "Om   w   H0   chi2"]))
```

```python
def plot_sct_hist(o_m,w,H0,burn_tail,title,save_title,hotext=[66,6000]):
    f, axarr = plt.subplots(2, 2,figsize=(12,8),gridspec_kw = {'width_ratios':[2,1],'height_ratios':[1,2]})
    f.text(0.5, 0.95, title,
         horizontalalignment='center',
         verticalalignment='top')
    axarr[0,0].hist(o_m[burn_tail:],bins=50)
    axarr[0,0].set_xlim([np.amin(o_m),np.amax(o_m)])
    
    axarr[1,0].scatter(o_m[burn_tail:],w[burn_tail:])
    axarr[1,0].set_xlim([np.amin(o_m),np.amax(o_m)])
    axarr[1,0].set_ylim([np.amin(w),np.amax(w)])
    axarr[1,0].set_ylabel(r'$w$')
    axarr[1,0].set_xlabel(r'$\Omega_M$')
    
    axarr[0,1].hist(H0[burn_tail:],bins=50)
    axarr[0,1].set_xlim([np.amin(H0), np.amax(H0)])
    axarr[0,1].text(hotext[0],hotext[1],r'$H_0$',fontsize=12)
    
    axarr[1,1].hist(w[burn_tail:],bins=50,orientation=u'horizontal')
    axarr[1,1].set_ylim([np.amin(w),np.amax(w)])
    
    plt.setp(axarr[0, 0].get_xticklabels(), visible=False)
    plt.setp(axarr[1, 1].get_yticklabels(), visible=False)
    
    plt.savefig('mcmc_'+save_title+'.pdf')
    plt.show()


def plot_conv_mcmc(o_m,w,H0,save_title):
    f, axarr = plt.subplots(3,1,figsize=(8,12))
    f.text(0.5, 0.95, 'Convergence',
         horizontalalignment='center',
         verticalalignment='top')
    
    axarr[0].plot(o_m)
    axarr[0].set_ylim([np.amin(o_m),np.amax(o_m)])
    axarr[0].set_ylabel(r'$\Omega_M$')
    
    axarr[1].plot(w)
    axarr[1].set_ylim([np.amin(w),np.amax(w)])
    axarr[1].set_ylabel(r'$w$')
    
    axarr[2].plot(H0)
    axarr[2].set_ylim([np.amin(H0),np.amax(H0)])
    axarr[2].set_ylabel(r'$H_0$')
    
    plt.setp(axarr[0].get_xticklabels(), visible=False)
    plt.setp(axarr[1].get_xticklabels(), visible=False)
    
    plt.savefig('mcmc_convergence_'+save_title+'.pdf')
    plt.show()
```

```python
#Get the 2D histagram
#Hist_mcmc will have #om lines and #w columns 
Hist_mcmc, omedges,wedges = np.histogram2d(o_m,w,bins=(40, 50))
om_c=np.array([(omedges[i]+omedges[i+1])/2 for i in range(len(omedges)-1)])
w_c=np.array([(wedges[i]+wedges[i+1])/2 for i in range(len(wedges)-1)])

#find 'most likelly' values
index_max=np.argwhere(Hist_mcmc==np.max(Hist_mcmc))
om_max=om_c[index_max[0,0]]
w_max=w_c[index_max[0,1]]
print(om_max,w_max)
```

```python
#find the 95% and 68% level

def confidence_levels(Hist,N):
    H2d=Hist.ravel()
    H2d = np.sort(H2d)
    tot=np.cumsum(H2d)
    Nmax=H2d[-1]
    N95=H2d[np.argwhere(tot>0.05*N)[0,0]]
    N68=H2d[np.argwhere(tot>0.32*N)[0,0]]
    return Nmax,N68,N95

#Make the contour plot

def contour_mcmc(Nmax,N68,N95,om_c,w_c,Hist_mcmc,om_max,w_max,save_title):
    import matplotlib as mpl
    colors = [(0.0, 0.0, 0.0, 0.7),(0.0, 0.0, 0.0, 0.4)] #black different alphas
    cmap = mpl.colors.ListedColormap(colors)
    
    levels=[N95,N68,Nmax]
    CS=plt.contourf(om_c,w_c, Hist_mcmc.T,levels=levels, cmap=cmap)
    plt.plot(om_max,w_max,'rx',ms=10)
    
    plt.xlim([np.amin(om_c),np.amax(om_c)])
    plt.ylim([np.amin(w_c),np.amax(w_c)])
    plt.xlabel(r'$\Omega_M$')
    plt.ylabel(r'$w$')
    
    plt.savefig('contour_om_w_'+save_title+'.pdf')
    plt.show()
```

```python
Nmax,N68,N95=confidence_levels(Hist_mcmc,N)
contour_mcmc(Nmax,N68,N95,om_c,w_c,Hist_mcmc,om_max,w_max,save_title)
```

```python
#in case we want something very smooth
x=[]
y=[]
z=[]
for i in range(len(om_c)):
    for j in range(len(w_c)):
        x.append(om_c[i])
        y.append(w_c[j])
        z.append(Hist_mcmc[i, j])


spl_red_2d = bisplrep(x, y,z,s=N+np.sqrt(2*N))
X = np.linspace(min(omedges), max(omedges), 100)
Y = np.linspace(min(wedges), max(wedges), 100)
Z = bisplev(X, Y, spl_red_2d)

CS=plt.contourf(X,Y,Z.T,levels=levels, cmap=cmap)
plt.plot(om_max,w_max,'rx',ms=10)
plt.xlim([0,0.5])
plt.ylim([-1.7,-0.5])
plt.xlabel(r'$\Omega_M$')
plt.ylabel(r'$w$')

plt.savefig('contour_om_w_union2.1_smooth_data3.pdf')
plt.show()
```

```python
H0_fid=67.74
w_fid=-1
Om_fid=0.31
Np=1000

# simulate z
z_data_sim=np.random.rand(Np)*(1.415-0.015)+0.015
z_data_sim.sort()
mu_t_sim=np.array([mu(z_data_sim_i,Om_fid,H0_fid,w_fid) for z_data_sim_i in z_data_sim])
mu_sim=np.zeros(Np)
for i in range(Np):
    mu_sim[i]=mu_t_sim[i]+np.random.randn()*0.3
err_mu_sim=0.3*np.ones(Np)

plt.figure(figsize = (12,6))
plt.plot(z_data_sim,mu_t_sim,'r',lw=2)
plt.errorbar(z_data_sim,mu_sim,yerr=[err_mu_sim,err_mu_sim],fmt='ko',ms=2,alpha=0.5)
plt.xlabel(r'$z$')
plt.ylabel(r'$\mu$')
plt.show()
plt.plot(z_data_sim,mu_t_sim-mu_sim,'ro',lw=2)
plt.xlabel(r'$z$')
plt.ylabel(r'$\Delta\mu$')
plt.show()
```

```python
o_m_sim,w_sim,H0_sim,chi2_mcmc_sim,acptr_sim=MCMC_SNe(100000,z_data_sim,mu_sim,0.3*np.ones(Np))
print(acptr_sim)
np.savetxt('mcmc_data_sim.txt',np.c_[o_m_sim,w_sim,H0_sim,chi2_mcmc_sim],fmt=['%2.3f','%2.3f','%2.3f','%2.3e'],\
           header='\n'.join(["acceptance rate %.2f" %acptr, "Om   w   H0   chi2"]))
```
